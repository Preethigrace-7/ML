Applying **Linear Regression** algorithm to **predict the salary** for employees in a particular comapany by taking the data which was given by the company for previously working professionals.

1.  Numpy

**Numpy** is fundamental library package for mathematical and scientific computing in Python.

**Matplotlib** is comprehensive library for creating static, animated and interactive visualizations in Python.

**Pandas** is an open-source library offers data structures and operations. It is used to import the data into our workspace.

```python
import numpy as n
import matplotlib.pyplot as pt
import pandas as p
```

here we inserting the data file which includes the data related to salary and year of experience

```python
dataset= p.read_csv('/content/salary-experience-dataset.csv')
x = dataset.iloc[:,0].values
y = dataset.iloc[:,1].values
```

```python
print(x)
```

```python
print(y)
```

```python
print(x.shape)
```

```python
print(y.shape)
```

Converting every data into an array

```python
x = n.reshape(x,(-1,1))
y = n.reshape(y,(-1,1))
print(x.shape)
print(y.shape)
print(x)
```

```python
print(y)
```

Plotting the data in graph format using matplotlib library

```python
pt.scatter(x,y,color = 'green') 
pt.title('salary vs experience') 
pt.xlabel('years of experience') 
pt.ylabel('salary')
pt. show()
```

```python
from sklearn.linear_model import LinearRegression 
regressor_l = LinearRegression() 
regressor_l.fit(x,y)
```

```python
pt.scatter(x,y,color = 'green') 
pt.plot(x,regressor_l.predict(x),color = 'blue') 
pt.title('salary vs experience') 
pt.xlabel('years of experience') 
pt.ylabel('salary')
pt.show()


```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=1/3, random_state=0) 
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape) 
print(y_test.shape)
```

```python
pt.scatter(x_train, y_train, color = 'green')
pt.scatter(x_test, y_test, color = 'yellow')
pt.title('salary vs experience') 
pt.xlabel('years of experience') 
pt.ylabel('salary')
pt.show()
```

```python
regressor_2 = LinearRegression() 
regressor_2.fit(x_train,y_train) 
y_pred = regressor_2.predict(x_test)
pt.scatter(x_test, y_test, color= 'green') 
pt.scatter(x_test, y_pred, color= 'yellow') 
pt.plot(x_test, y_pred, color= 'black') 
pt.title('salary vs experience') 
pt.xlabel('years of experience') 
pt.ylabel('salary')
pt.show()
```

MSE (Mean Squared Error) or Mean Squared Deviation means the average aquared difference between the estimated value and error value.

MAE (Mean Absolute Error) means the average difference between the calculated values and the actual values.

```python
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error 
print('MSE', mean_squared_error(y_test, y_pred)) 
print('MAE', mean_absolute_error(y_test, y_pred))
```
